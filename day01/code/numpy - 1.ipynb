### 4. 级联
1. np.concatenate()
级联需要注意的点：
- 级联的参数是列表：一定要加中括号或小括号
- 维度必须相同
- 形状相符
- 【重点】级联的方向默认是shape这个tuple的第一个值所代表的维度方向
- 可通过axis参数改变级联的方向

```python
n1 = np.random.randint(100, size=(4,5))
n2 = np.random.randint(100, size=(4,5))
display(n1,n2)
```

```python
# axis=0代表行方向上的级联.
np.concatenate([n1,n2], axis=0)
```

```python
# axis=1代表列方向上的级联.
np.concatenate((n1,n2), axis=1)
```

```python
n2 = np.random.randint(100,size=(2,5))
n2
```

```python
# 进行行级联的时候,列数必须相等.
np.concatenate((n1,n2))
```

```python
# 进行列级联,行数必须相等.
n3 = np.random.randint(100, size=(4, 7))
n3
```

```python
np.concatenate((n1,n3), axis=1)
```

- 规则一：为缺失的维度补1
- 规则二：假定缺失元素用已有值填充


例1：
m = np.ones((2, 3))
a = np.arange(3)
求M+a

```python
m  = np.ones((2,3))
a = np.arange(3)
display(m,a)
```

```python
m + a
# 给a补一个维度
#[[0, 1, 2]]
# 给a用已有的元素去填充缺失 的值
# [[0, 1, 2],
    [0, 1, 2]]
# 然后再对应的元素相加
```

## 六、ndarray的排序

小测验：
使用以上所学numpy的知识，对一个ndarray对象进行选择排序。  

def Sortn(x):  

代码越短越好

```python
# 什么是选择排序
# ,然后和当前元素进行替换.
def select(arr):
    for i in range(len(arr) -1):
        # 找到待排序数组中最小的
        index = np.argmin(arr[i:]) + i
        # ,然后和当前元素进行替换.
        if arr[i] > arr[index]:
            arr[i], arr[index] = arr[index], arr[i]
    return arr
```

```python
a = np.array([1,2,5,3,6,])
```

```python
select(a)
```
